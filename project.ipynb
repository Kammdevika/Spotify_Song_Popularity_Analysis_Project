<h1><b>Spotify Song Popularity</b></h1>

<h3><b> Objective</b></h3>

<h5>To analyze Spotify song data to uncover factors influencing song popularity and identify trends in popular tracks.</h5>

<h3><b>Importing Libarires</b></h3>

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

<h3><b> Data Reading </b> </h3>

```python
#reading data using pandas libararies
df = pd.read_csv("C:/Users/Devika/Downloads/universal_top_spotify_songs.csv")
df.head()
```

<h3><b> Data Cleaning</b></h3>

```python
#checks null value for each column 
pd.isnull(df).sum()
```

```python
#drop 
df.dropna(inplace=True)
df
```

```python
pd.isnull(df).sum()
```

```python
#info
df.info()
```

```python
#using describe() method to findout the statical(min, max, std, mean, count, 25%, 75%)
df.describe().transpose()
```

```python
#sorting the data, finding the top 10 song through the popularity col
sort_df = df.sort_values('popularity', ascending = False).head(10)

# Display the sorted data with selected columns
sort_df[['name', 'artists', 'popularity','country']]
```

```python
# Filter songs with Popularity > 90 and sort by Popularity in ascending order
filter_df = df.query('popularity > 90').sort_values('popularity', ascending=True)

# Display the filtered data with selected columns
filter_df[['name', 'artists', 'popularity','country']]
```

```python
# Create a new column for duration in minutes
df['duration (minutes)'] = df['duration_ms'] / 60000

# Display the first few rows to verify
print(df[['duration_ms', 'duration (minutes)']].head())
```

<h3><b> Data Visualization</b></h3>

```python
# Group by 'Artist Name(s)' and get the mean popularity for each artist
artist_popularity = df.groupby('artists')['popularity'].mean().sort_values(ascending=False)

# Plot the bar chart
plt.figure(figsize=(10, 6))
artist_popularity.head(10).plot(kind='barh', color='salmon')  # Top 10 artists for popularity
plt.title('Top 10 Artists vs Average Popularity')
plt.xlabel('Average Popularity')
plt.ylabel('Artist Names')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
#popularity vs danceability
plt.figure(figsize=(10, 6))
sns.scatterplot(data=df, x='danceability', y='popularity', hue='danceability', palette='viridis')
plt.title('Popularity vs Danceability (Scatter Plot)')
plt.xlabel('Danceability')
plt.ylabel('Popularity')
plt.tight_layout()
plt.show()
```

```python
# Select numerical columns for correlation analysis
numerical_cols = ['popularity', 'danceability', 'energy', 'loudness', 
                  'valence', 'tempo', 'speechiness', 'acousticness', 
                  'instrumentalness', 'liveness']

# Calculate the correlation matrix
corr_matrix = df[numerical_cols].corr()

# Plot the heatmap
plt.figure(figsize=(10, 8))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt='.2f', linewidths=0.5)
plt.title('Correlation Heatmap')
plt.tight_layout()
plt.show()
```

```python
#hexbin plot for song duration(minutes) vs popularity
plt.figure(figsize=(12, 6))
plt.hexbin(df['duration (minutes)'], df['popularity'], gridsize=30, cmap='Reds', mincnt=1)
plt.colorbar(label='Count')
plt.title('Song Duration vs. Popularity (Hexbin Plot)')
plt.xlabel('Duration (minutes)')
plt.ylabel('Popularity')
plt.show()
```

```python
# Create a density plot
plt.figure(figsize=(10, 6))
sns.kdeplot(df['popularity'], fill=True, color='indigo')
plt.title('Density Plot of Popularity')
plt.xlabel('Popularity')
plt.ylabel('Density')
plt.tight_layout()
plt.show()
```

```python
# Scatter plot of Danceability vs Popularity
plt.figure(figsize=(10, 6))
plt.scatter(df['danceability'], df['energy'], alpha=0.6, color='purple')
plt.title('Danceability vs Energy')
plt.xlabel('Danceability')
plt.ylabel('Energy')
plt.tight_layout()
plt.show()
```

```python
# Group by 'Track Name' and get the average popularity, then sort by popularity
track_popularity = df.groupby('name')['popularity'].mean().sort_values(ascending=False)

# Plot the top 20 tracks by popularity
plt.figure(figsize=(10, 10))
track_popularity.head(20).plot(kind='bar', color='red')
plt.title('Top 20 Tracks vs Popularity')
plt.xlabel('Track Name')
plt.ylabel('Average Popularity')
plt.tight_layout()
plt.show()
```

```python
# Count the number of songs per country
country_song_count = df['country'].value_counts()

# Plot the pie chart
plt.figure(figsize=(8, 8))
country_song_count.head(10).plot(kind='pie', autopct='%1.1f%%', colors=sns.color_palette('Set3', n_colors=10))
plt.title('Top 10 Countries by Number of Songs')
plt.ylabel('')
plt.tight_layout()
plt.show()
```

```python
# Group by 'country' and calculate average popularity
country_popularity = df.groupby('country')['popularity'].mean().sort_values(ascending=False)

# Plot the bar chart
plt.figure(figsize=(12, 6))
country_popularity.plot(kind='bar', color='skyblue')
plt.title('Average Popularity by Country')
plt.xlabel('Country')
plt.ylabel('Average Popularity')
plt.tight_layout()
plt.show()
```
